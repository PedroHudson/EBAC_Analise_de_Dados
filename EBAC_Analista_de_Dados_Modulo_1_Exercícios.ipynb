**2º Exercício: Números**

Preencha as células de código para incluir os valores de (A), (B) e (C) na tabela de ticket médio.

<br>

| Dia   | Valor Total Vendas | Qtd Total Vendas | Ticket Medio |
|-------|--------------------|------------------|--------------|
| 19/01 | (A)                | 3                | 320.52       |
| 20/01 | 834.47             | (B)              | 119.21       |
| 23/01 | 15378.12           | 5                | (C)          |

```python
Qtd_Total_Vendas_A = 3
Ticket_Medio_A = 320.52
Valor_Total_Vendas_A = Qtd_Total_Vendas_A * Ticket_Medio_A
A = Valor_Total_Vendas_A
print (A)
```

```python
Valor_Total_Vendas_B = 834.47
Ticket_Medio_B = 119.21
Qtd_Total_Vendas_B = Valor_Total_Vendas_B / Ticket_Medio_B
B = Qtd_Total_Vendas_B
print (int(B))
```

```python
Valor_Total_Vendas_C = 15378.12
Qtd_Total_Vendas_C = 5
Ticket_Medio_C = Valor_Total_Vendas_C / Qtd_Total_Vendas_C
C = Ticket_Medio_C
print (C)
```

**3º Exercício: String**

Aplique três métodos distintos na string indicada.

**Parte 2**

Extraia da string indicada o valor da taxa selic na variável selic e o valor do ano na variável ano . Imprima os valores na tela.

```python
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'

posicao_selic = noticia.find('%')
print (posicao_selic)

selic = noticia[12:17]
print (selic)

posicao_anos = noticia.find('.')
print (posicao_anos)

ano = noticia[62:68]
print (ano)
```
